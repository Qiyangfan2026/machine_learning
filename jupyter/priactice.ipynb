```python
import numpy as np
import matplotlib.pyplot as plt

# 生成模拟数据（y = 2*x + 1 + 噪声）
np.random.seed(30)
X = np.random.rand(100, 1) * 10          # 100个样本，特征值 0-10
y = 2 * X + 1 + np.random.randn(100, 1)  # 真实标签，加上噪声

# 添加偏置项（x0 = 1）
X_b = np.c_[np.ones((100, 1)), X]        # shape: (100, 2)

# 初始化参数（theta0, theta1）
theta = np.random.randn(2, 1)

# 超参数
learning_rate = 0.001
n_iterations = 1000
tiduxiajia
# 梯度下降
for iteration in range(n_iterations):
    gradients = 2/len(X_b) * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients

print("拟合结果：", theta.flatten())   # 应接近 [1, 2]

# 画图验证
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta), color='red')
plt.show()
```
